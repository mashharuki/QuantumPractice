# 量子テレポーテーション（シミュレータ）
アリスからボブへキュビット０の状態を転送

## ライブラリのインポート

```python
from qiskit import BasicAer, execute
from qiskit import QuantumRegister, ClassicalRegister, QuantumCircuit
from qiskit.quantum_info import Statevector
from qiskit.visualization import plot_histogram
from qiskit.visualization import plot_bloch_multivector
from numpy import pi
```

## 量子回路
３つのキュビットとビットを用意する

```python
# ３つのキュビットを用意する。
qreg = QuantumRegister(3, 'q')
# ３つのクラシカルビットにラベルをつける。
c0 = ClassicalRegister(1, 'c0')
c1 = ClassicalRegister(1, 'c1')
c2 = ClassicalRegister(1, 'c2')
# 回路を用意する。
qc = QuantumCircuit(qreg, c0, c1, c2)
```

アリスがこれから転送するキュビット０用意

```python
# 60だけずらす。
# RYゲートを追加する。
qc.ry(pi/3, 0)
# 区切る
qc.barrier(0,1,2);
```

回路の確認

```python
qc.draw(output='mpl')
```

ブロッフォ球で確認

```python
state = Statevector.from_instruction(qc)
plot_bloch_multivector(state)
```

確立（理論値）表示

```python
counts = state.probabilities_dict()
plot_histogram(counts)
```

第三者がキュビット１とキュビット２をもつれさせ、キュビッチ１をアリスへ渡し、キュビット２をボブに渡したことにする

```python
# HADゲートを追加する。(重ね合わせ)
qc.h(1)
# CNOTゲートを追加する。
qc.cx(1,2)
qc.barrier(0,1,2)
qc.draw(output='mpl')
```

アリスがキュビット０の状態を送信

```python
# キュビット0と1をもつれ状態にする。
qc.cx(0,1)
qc.h(0)
# 区切る
qc.barrier(0,1,2)
# キュビット0と1を測定する。
qc.measure(0,0)
qc.measure(1,1)
qc.barrier(0,1,2)

qc.draw(output='mpl')
```

ボブが、アリスから測定結果を受信し、第三者からあらかじめ得たキュビット２を制御（調整）.キュビット２は、アリスが用意したキュビット０と同じ状態となる.

```python
# Zゲート(位相反転)を追加する。 (クラシカルビット０が1なら反転させる)
qc.z(2).c_if(c0, 1)
#  NOTゲートを追加する。 (クラシカルビット1が1なら反転させる)
qc.x(2).c_if(c1, 1)

qc.draw(output='mpl')
```

```python
statevector_backend = BasicAer.get_backend('statevector_simulator') 
result = execute(qc, statevector_backend).result()
final_state = result.get_statevector()
plot_bloch_multivector(final_state)
```

## 確率ヒストグラムの表示
キュビット２も測定し、ビットパターン毎の確率を棒グラフで表示.

```python
qc.measure(2,2)

qc.draw(output='mpl')
```

```python
qasm_backend = BasicAer.get_backend('qasm_simulator') 
result = execute(qc, qasm_backend, shots=1000).result()
final_counts = result.get_counts()
```

```python
plot_histogram(final_counts)
```
